```python
import cv2
import numpy as np

cap = cv2.VideoCapture('/media/amiyaun/New Volume/rugved/task3/Ball_Tracking.mp4')

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/media/amiyaun/New Volume/rugved/task3/Tracked_Ball_Output.mp4',fourcc,20.0,(640, 480))


lower_color = np.array([35, 100, 100])
upper_color = np.array([80, 240, 240])

while cap.isOpened():
    ret, frame = cap.read()
    if ret == False:
        break
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    mask = cv2.inRange(frame, lower_color, upper_color)
    contours = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        (x, y), radius = cv2.minEnclosingCircle(largest_contour)
        if radius > 100:
            cv2.circle(frame, (x, y), radius, (0, 255, 0), 2)
            cv2.circle(frame, (x, y), 5, (0, 0, 255), -1)

    out.write(frame)

cap.release()
out.release()

from google.colab import files
files.download('/content/Tracked_Ball_Output.mp4')
```
